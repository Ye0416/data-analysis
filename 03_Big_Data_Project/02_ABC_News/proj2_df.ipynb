```python
from pyspark.sql.session import SparkSession
from pyspark.sql.functions import *
import sys
import re
```

```python
spark = SparkSession.builder.master("local").appName("proj2").getOrCreate()
```

```python
fileDF = spark.sparkContext.textFile('pg100.txt')
```

```python
words = fileDF.flatMap(lambda line: re.split("[\\s*$&#/\"'\\,.:;?!\\[\\](){}<>~\\-_]+", line))  
wordsDF1 = words.map(lambda x: (x, ))
wordsDF1.collect()
```

```python
wordsDF2 = wordsDF1.toDF("word string")
wordsDF2.collect()           # toDF(' ')
```

```python
wordsDF3 = wordsDF2.withColumn("word", lower(col("word")))
wordsDF3.collect()                # .withColumn(,)
```

```python
wordsDF4 = wordsDF3.filter(length(col("word")) >=1).filter((col("word").substr(0,1)<= 'z') & (col("word").substr(0,1)>='a'))
wordsDF4.collect()          # fileter, 
```
